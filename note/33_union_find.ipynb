            root = parents[root]
        }

        return root;
    }


    public void union(int p, int q){
        int proot = findRoot(p);
        int qroot = findRoot(q);
        if proot == qroot:
            return 

        // p  q 
        if proot.rank < qroot.rank
            proot.parent = qroot; // rankrank
        else if proot.rank > qroot.rank
            qroot.parent = proot;
        else
            qroot.parent = proot;
            proot.rank +=1; // rankqpprank+1
    }
}

public class QuickUnionUF(){
    private int[] parents;

    public QuickUnionUF(int n){
        parents = new int[n];
        for (int i=0;i<N;i++){
            parents[i] = i; // parent
        }
    }

    private int findRoot(int i){
        int root = i;
        while (root != parents[root])
            root = parents[root];

        while(i != parents[i]){ // 
            int tmp = parents[i]; 
            parents[i] = root; 
            i = tmp;
        }
    }

    public void union(int p, int q){
        int proot = findRoot(p);
        int qroot = findRoot(q);
        parents[proot] = qroot;
    }
}
